```python
import folium
import pandas as pd
# Philadephia latitude and longitude values
latitude = 39.95233
longitude = -75.16379

# Create map and display it
phi_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# Display the map
phi_map
```

```python
# Read Dataset 
agencies = pd.read_csv('HousingCounselingAgencies.csv')
agencies.head()
```

```python
# get the first 200 housing counceling agencies in the dataset
limit = 200
data = agencies.iloc[0:limit, :]

# Instantiate a feature group for the housing counceling agencies in the dataframe
housing = folium.map.FeatureGroup()

# Loop through the agencies and add each to the incidents feature group
for lat, lng, in zip(agencies.Y, data.X):
    housing.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )

# Add housing to map
phi_map = folium.Map(location=[latitude, longitude], zoom_start=12)
phi_map.add_child(housing)
```

```python
# add pop-up text to each marker on the map
latitudes = list(data.Y)
longitudes = list(data.X)
labels = list(data.AGENCY)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(phi_map)    
    
# add housing to map
phi_map.add_child(housing)
```
